```python
import statsmodels.api as sm

import numpy as np
import pandas as pd

```

```python
data = pd.read_excel('Day_Mids.xlsx')
data.head()
```

```python
y = data.iloc[:,0]
X = sm.add_constant(data.iloc[:,1:])
X.head()
```

```python
model = sm.OLS(y, X).fit()

results = pd.DataFrame(columns=['Intercept','Beta_1Day','Beta_2Day','Beta_3Day','Beta_4Day',  'P-Value Intercept', 'P-Value Beta_1Day', 'P-Value Beta_2Day', 'P-Value Beta_3Day', 'P-Value Beta_4Day'])
results.loc[0] = [model.params[0],model.params[1],model.params[2],model.params[3],model.params[4], model.pvalues[0], model.pvalues[1], model.pvalues[2], model.pvalues[3], model.pvalues[4]]

results
```
